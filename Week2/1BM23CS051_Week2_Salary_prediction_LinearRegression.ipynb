```python
df['YearsExperience'].fillna(df['YearsExperience'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
q1=df['YearsExperience'].quantile(0.25)
q3=df['YearsExperience'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
df_clean = df[(df['YearsExperience'] >= lower_bound) & (df['YearsExperience'] <= upper_bound)]
df_clean.head()
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
model=LinearRegression()
X=df[['YearsExperience']]
Y=df['Salary']
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state=42)
model.fit(xtrain,ytrain)
```

```python
predicted_salary = model.predict([[12]])
print("Predicted Salary for 12 years experience:", predicted_salary[0])
```
